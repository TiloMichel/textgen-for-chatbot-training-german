```python
url = 'https://www.rki.de/SharedDocs/FAQ/NCOV2019/gesamt.html' #@param ["https://kmi.fbi.h-da.de/aboutkmi.html", "https://www.rki.de/SharedDocs/FAQ/NCOV2019/gesamt.html", "https://www.dge.de/index.php?id=1570"]

import requests

response = requests.get(url)
print(response.encoding)
text = response.text
if response.encoding == "ISO-8859-1":
    text = response.text.encode('latin1').decode('utf8')
```

jusText with requests https://github.com/miso-belica/jusText

```python
import justext

paragraphs = justext.justext(text, justext.get_stoplist("German"))
for paragraph in paragraphs:
  if not paragraph.is_boilerplate:
    print(paragraph.text)
```

Trafilatura https://github.com/adbar/trafilatura

Example from https://trafilatura.readthedocs.io/en/latest/usage-python.html#quickstart

```python
from trafilatura import fetch_url, extract

downloaded = fetch_url(url)
result = extract(downloaded)
print(result)
```
